```python
import yfinance as yf

# Fetch Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset the index
tesla_data_reset = tesla_data.reset_index()

# Save the data to a CSV file
tesla_data_reset.to_csv('tesla_stock_data.csv', index=False)

# Display the first five rows
print(tesla_data_reset.head())
```

```python
import pandas as pd
import yfinance as yf

# Fetch Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Add a revenue column (this is hypothetical since exact revenue data isn't directly available from yfinance)
# Here we'll just use volume * close price as a proxy for revenue
tesla_data['Revenue'] = tesla_data['Volume'] * tesla_data['Close']

# Reset index to make Date a column
tesla_revenue = tesla_data.reset_index()

# Display the last five rows
print(tesla_revenue.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset the index to make Date a column for plotting
tesla_data_reset = tesla_data.reset_index()

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(15, 8))

# Plot the closing price
ax.plot(tesla_data_reset['Date'], tesla_data_reset['Close'], label='Close Price')

# Set the title of the graph
plt.title("Tesla Stock Price Over Time")

# Label the axes
ax.set_xlabel('Date')
ax.set_ylabel('Stock Price (USD)')

# Add a legend for the plot
ax.legend()

# Rotate x-axis labels so they don't overlap
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
